```python
import pandas as pd
```

```python
jan = pd.read_csv('DEMACRO/2020/jan2020.csv', sep = ';')
fev = pd.read_csv('DEMACRO/2020/fev2020.csv', sep = ';')
mar = pd.read_csv('DEMACRO/2020/mar2020.csv', sep = ';')
abr = pd.read_csv('DEMACRO/2020/abr2020.csv', sep = ';')
mai = pd.read_csv('DEMACRO/2020/mai2020.csv', sep = ';')
jun = pd.read_csv('DEMACRO/2020/jun2020.csv', sep = ';')
jul = pd.read_csv('DEMACRO/2020/jul2020.csv', sep = ';')
```

```python
todos = pd.concat([jan, fev, mar, abr, mai, jun, jul])
```

```python
todos.head(100)
```

```python
todos.drop_duplicates(subset ="NUM_BO", keep = False, inplace = True)
```

```python
todos = todos.drop(['DELEGACIA_NOME', 'DELEGACIA_CIRCUNSCRICAO', 'ESPECIE', 'RUBRICA'], axis = 1)
```
