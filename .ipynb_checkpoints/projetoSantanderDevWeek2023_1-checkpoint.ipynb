# extract

```python
import pandas as pd
```

```python
cap_2018 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_capital_2018.csv', sep = ',')
cap_2019 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_capital_2019.csv', sep = ',')
cap_2020 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_capital_2020.csv', sep = ',')
met_2018 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_metro_2018.csv', sep = ',')
met_2019 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_metro_2019.csv', sep = ',')
met_2020 = pd.read_csv('projetoSantanderDevWeek2023/DADOS/BASES_COMPILADAS/furtos_metro_2020.csv', sep = ',')
```

```python
todos = pd.concat([cap_2018, cap_2019, cap_2020, met_2018, met_2019, met_2020])
```

# Transform

#### Deletando colunas com dados inrrelevantes e tratando os nulos

```python
todos1 = todos.drop(["HORAOCORRENCIA","LATITUDE","LONGITUDE"], axis = 1)
```

```python
todos1[['LOGRADOURO']]=todos1[['LOGRADOURO']].fillna('')
todos1[['BAIRRO']]=todos1[['BAIRRO']].fillna('')
todos1[['DESCR_COR_VEICULO']]=todos1[['DESCR_COR_VEICULO']].fillna('')
todos1[['DESCR_MARCA_VEICULO']]=todos1[['DESCR_MARCA_VEICULO']].fillna('')
todos1[['ANO_FABRICACAO']]=todos1[['ANO_FABRICACAO']].fillna('')
todos1[['ANO_MODELO']]=todos1[['ANO_MODELO']].fillna('')
todos1[['DESCR_TIPO_VEICULO']]=todos1[['DESCR_TIPO_VEICULO']].fillna('')
print(todos1.isnull().sum())
```

```python
todos1.tail()
```

### Passando datas para formato Datetime() e núneros para int()

```python
todos1['BO_INICIADO'] = pd.to_datetime(todos1['BO_INICIADO'])
todos1['BO_EMITIDO'] = pd.to_datetime(todos1['BO_EMITIDO'])
todos1['DATACOMUNICACAO'] = pd.to_datetime(todos1['DATACOMUNICACAO'])
todos1['DATAOCORRENCIA'] = pd.to_datetime(todos1['DATAOCORRENCIA'], errors="ignore")                                            
```

```python
convert_dict = {'ANO_BO':int,'NUMERO':int}
todos1 = todos1.astype(convert_dict)


print(todos1.dtypes)
```
